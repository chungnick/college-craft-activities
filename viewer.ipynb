```python
import pandas as pd

csv_path = "/Users/nick/college-craft-extracurriculars/ec_bank_rows_with_valid_url.csv"
df = pd.read_csv(csv_path)

# Confirm valid_url is filled for all rows
filled_mask = df["valid_url"].astype(str).str.strip().ne("") & df["valid_url"].notna()
print(f"Rows: {len(df)}")
print(f"valid_url filled: {filled_mask.all()} ({filled_mask.sum()}/{len(df)})")
print(df["valid_url"].value_counts(dropna=False).head(10))

assert filled_mask.all(), "Some rows have missing/empty valid_url"
```

```python
# show rows with valid_url and non null values for application_due_dates
filtered_df = df[filled_mask & df["application_due_dates"].notna()]
print(filtered_df)
```
